### 1. What is a random variable in probability theory?

- In probability theory, a random variable is a variable that takes on numerical values determined by the outcome of a random phenomenon.

- Two Key Types of Random Variables:

  1. Discrete Random Variable: Takes on countable values ex- integers. ex- number rolled on a fair six-sided die (values: 1, 2, 3, 4, 5, 6).

  2. Continuous Random Variable: Takes on values from a continuous range ex- all real numbers in an interval. ex- Exact height of a randomly selected person (could be any value within a range, like 150.5 cm, 170.1 cm etc.

- Why Use Random Variables?

  1. Quantify uncertainty.

  2. Use mathematical tools like probability distributions, expectation, variance to analyze random outcomes.

### 3. What is the difference between discrete and continuous distributions?

- Difference between discrete and continuous distributions lies in the type of values the random variable can take and how probabilities are assigned.

  1. Discrete Distribution
  
    - Discrete distribution describes the probability of outcomes of a discrete random variable, which can take countable values (finite or countably infinite).

    - Values are distinct and separate ex- 0, 1, 2,....

    - Each value has a non-zero probability.

    - Probabilities are given by a probability mass function (PMF).

    - ex- Rolling a fair die

  2. Continuous Distribution

    - Continuous distribution describes the probability of outcomes of a continuous random variable, which can take any value within an interval (uncountably infinite).

    - Values can lie anywhere on a real interval ex- all numbers between 0 and 1).

    - probability at a single point is zero

    - Probabilities are given by a probability density function (PDF).

    - ex- Height of adults (approximate)

### 5. How do cumulative distribution functions (CDF) differ from probability distribution functions (PDF)?

- Probability Density Function (PDF) and a Cumulative Distribution Function (CDF) are both used to describe the behavior of continuous random variables, but they serve different purposes.

> PDF

  1. PDF, denoted by f(x) represents the relative likelihood that the random variable takes a value near x.
  
  2. It does not give the probability of the variable being exactly equal to x since P(X = x) = 0 for continuous variables.
  
  3. PDF describes probability density at a point

  4. PDF is the derivative of the CDF

> CDF

  1. CDF, denoted by F(x) gives the cumulative probability that the random variable is less than or equal to a certain value x.
  
  2. CDF is a non-decreasing function that ranges from 0 to 1.

  3. CDF describes the total probability up to that point

  4. CDF is the integral of the PDF

### 7. What are the key properties of a Bernoulli distribution?

- Bernoulli distribution is a discrete probability distribution for a random variable

- It has only two possible outcomes: success (usually denoted by 1) and failure (denoted by 0).

- It is the simplest form of a discrete distribution and is used to model a single trial of a binary experiment.

> Key properties

  1. Possible Values: X ‚àà {0,1}

  2. Probability Mass Function (PMF): P(X = x) {p if x= 1, 1-p if x= 0}

  3. Mean (Expected Value): E[X] = p

  4. Variance: Var(X) = p (1-p)

  5. Moment Generating Function (MGF)

  6. Applications:

    - Modeling binary outcomes ex- coin toss, yes/no survey response.

    - Building block for the binomial distribution.

- Bernoulli distribution is fundamental in probability theory and statistics and forms the basis of more complex discrete distributions.

### 8. What is the binomial distribution, and how is it used in probability?

- Binomial Distribution is a discrete probability distribution that models the number of successes in a fixed number of independent and identical Bernoulli trials, each with the same probability of success.

- It is used when there are only two possible outcomes in each trial: success (with probability p) and failure (with probability 1-p)

- Probability Mass Function (PMF): If a random variable X follows a binomial distribution with parameters n (number of trials) and p (probability of success)

> Key Properties

  1. Mean (Expected Value): E[X] = np

  2. Variance: Var(X) = np(1-p)

  3. Support: X ‚àà {0,1,2,...,n}

> Uses in Probability: binomial distribution is used to model

  1. number of heads in n coin tosses.

  2. number of defective items in a batch.

  3. number of correct answers in a multiple-choice test (with guessing).

- It is widely applied in quality control, survey analysis, and risk assessment where outcomes are binary and trials are independent.

### 13.  What is the Central Limit Theorem (CLT), and why is it critical in statistics?

- Central Limit Theorem (CLT) states that the sampling distribution of the sample mean approaches a normal distribution as the sample size increases, regardless of the shape of the population distribution, provided the samples are independent and identically distributed.

> Why it is critical:

  1. It allows for normal-based inference (confidence intervals, hypothesis tests) even when the population is not normally distributed.

  2. It underpins many statistical procedures and models.
