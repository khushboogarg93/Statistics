### 1. What is hypothesis testing in statistics?

- Hypothesis testing in statistics is a formal procedure used to evaluate assumptions or claims (hypotheses) about a population parameter based on sample data.

- It involves making a decision between two competing hypotheses:

  1. Null Hypothesis (H0) - A statement that there is no effect, no difference, or no relationship. It represents the status quo or default assumption.

  2. Alternative Hypothesis (H1 or Ha) - A statement that contradicts the null hypothesis, suggesting an effect, difference, or relationship exists.

- Goal of hypothesis testing is to determine whether the observed data provides sufficient evidence to reject the null hypothesis in favor of the alternative. This decision is based on:

  1. Significance level (alpha) - The probability threshold (commonly 0.05) for rejecting H0.

  2. Test statistic: A value computed from sample data that is compared against a theoretical distribution.

  3. P-value: The probability of obtaining a result at least as extreme as the observed one, assuming H0 is true.

- If the p-value is less than or equal to α, the null hypothesis is rejected; otherwise, it is not rejected. Hypothesis testing is a foundational tool in inferential statistics for drawing conclusions from data.

### 2. What is the null hypothesis, and how does it differ from the alternative hypothesis?

- Null hypothesis and the alternative hypothesis are two opposing statements used in hypothesis testing.

  1. Null Hypothesis (H0) -

    - It is a statement of no effect, no difference, or no relationship.

    - It represents the default or assumed condition.

    - The purpose is to test whether there is enough evidence to reject this assumption.

    - Ex- The average height of students is 170 cm, H0 : μ = 170

  2. Alternative Hypothesis (H1 or Ha)

    - It is a statement that contradicts the null hypothesis.

    - It represents the claim or effect we are trying to find evidence for.

    - If there is sufficient evidence, the null hypothesis is rejected in favor of the alternative.

    - Ex- The average height of students is not 170 cm, H1 : μ != 170

- Hypothesis testing evaluates whether the data provides enough evidence to reject the null hypothesis and support the alternative hypothesis.

  3. Not the Probability that H0 is True:

### 6. What are Type 1 and Type 2 errors in hypothesis testing?

- In hypothesis testing, Type I and Type II errors are two kinds of incorrect decisions that can be made when testing a hypothesis.

> Type I Error (False Positive):

  1. Occurs when the null hypothesis (H0) is true, but it is incorrectly rejected.

  2. You conclude there is an effect or difference when there is not.

  3. probability of making a Type I error is denoted by α (significance level).

  4. Ex- A medical test incorrectly shows a patient has a disease when they actually don't.

> Type II Error (False Negative):

  1. Occurs when the null hypothesis (H0) is false, but it is incorrectly not rejected.

  2. You fail to detect an effect or difference that actually exists.

  3. probability of making a Type II error is denoted by β.

  4. Ex- A medical test fails to detect a disease in a patient who actually has it.

> Why They Matter:

  1. Minimizing Type I errors is important when false positives have serious consequences ex- criminal trials.

  2. Minimizing Type II errors is important when missing a true effect is costly ex- failing to detect a disease.

- Type I and Type II errors represent the two main risks in statistical decision-making, and balancing them is a key part of designing a sound hypothesis test.

### 7. What is the difference between a one-tailed and a two-tailed test in hypothesis testing?

- In hypothesis testing, the difference between a one-tailed and a two-tailed test lies in the direction of the alternative hypothesis and how the rejection region is determined.

> One-Tailed Test:

  1. A test that examines whether a parameter is either greater than or less than a specific value, but not both.

  2. Alternative Hypothesis (H1) - Specifies a directional claim. Ex- h1 : μ > μ0 or h1 : μ < μ0

  3. Rejection Region: Located in one tail of the distribution (either the left or the right).

  4. Use When: You are only interested in deviation in one direction.

  5. Use a one-tailed test when the research hypothesis predicts a specific direction.


> Two-Tailed Test:

  1. A test that examines whether a parameter is different (either higher or lower) from a specific value.

  2. Alternative Hypothesis (H1) - Specifies a non-directional claim. Ex- h1 : μ != μ0

  3. Rejection Region: Split between both tails of the distribution.

  4. Use When: You are interested in deviations in either direction.

  5. Use a two-tailed test when the hypothesis is only about a difference, regardless of direction.

### 9. How do you calculate the Z-score, and what does it represent in hypothesis testing?

- In hypothesis testing, the Z-score (or Z-value) is a standardized value that indicates how many standard deviations an observed sample statistic is from the population mean, assuming the null hypothesis is true.

> How to Calculate the Z-score:

  - For a one-sample Z-test, the formula is: Z = (X(bar) - μ) / (σ / Sqrt(n)

  - Where X(bar) = samoke mean, μ = population mean under the null hypothesis, σ= population standard deviation, n = sample size.

> Z-score Represents:

  1. The Z-score tells you how far, in standard deviation units, the sample mean is from the hypothesized population mean.

  2. It allows you to determine the probability (p-value) of observing a test statistic as extreme as the one calculated, under the assumption that the null hypothesis is true.

> Interpretation of Z-scores in Hypothesis Testing:

  1. Z-score near 0: Sample mean is close to the population mean → likely fail to reject (H0).

  2. large positive or negative Z-score: Sample mean is far from the population mean → may reject (H0). if it falls in the rejection region.

- Z-score is a key component in Z-tests and represents how many standard deviations a sample result is from the expected value under the null hypothesis.

- It helps assess whether the observed data is statistically significant.

  3. Paired t-test: Comparing means from the same group at different times.

- t-distribution is used instead of the normal distribution when dealing with small samples and/or an unknown population standard deviation.

### 12. What is the T-test, and how is it used in hypothesis testing?

- t-test is a statistical method used in hypothesis testing to determine whether there is a significant difference between means.

- It is especially useful when the population standard deviation is unknown and/or the sample size is small.

- It evaluates whether the observed difference in sample means (or a sample mean and a known value) is statistically significant.

> Types of T-tests:

  1. One-Sample T-test: Tests whether the mean of a single sample differs from a known or hypothesized population mean.

  2. Independent Two-Sample T-test: Compares the means of two independent groups to see if they are significantly different.

  3. Paired Sample T-test: Compares means from the same group at different times or under different conditions ex- before-and-after studies.

> How the T-test Is Used in Hypothesis Testing:

  1. State the Hypotheses

  2. Choose the Significance Level (α), typically 0.05

  3. Calculate the T-statistic using the appropriate formula for the test type.

  4. Determine the Degrees of Freedom and find the critical t-value from the t-distribution table.

  5. Compare the t-statistic to the critical value or use the p-value.

  6. Make a Decision: If the t-statistic is in the rejection region or p-value ≤ α, reject H0. Otherwise, fail to reject H0.

- t-test is a versatile and widely used tool in hypothesis testing, especially when dealing with small samples and unknown population standard deviations.

- It helps determine if observed differences in means are likely due to chance or represent a real effect.

### 13. What is the relationship between Z-test and T-test in hypothesis testing?

> Relationship Between Z-test and T-test:

### 14. What is a confidence interval, and how is it used to interpret statistical results?

- Confidence interval (CI) is a range of values, derived from sample data, that is used to estimate an unknown population parameter (such as a mean or proportion) with a certain level of confidence.

- It is expressed as: Point estimate ± Margin of error

- Confidence level (commonly 90%, 95%, or 99%) represents the degree of certainty that the interval contains the true population parameter.

- 95% confidence interval, for example, means that if we were to take many random samples and build a confidence interval from each, about 95% of those intervals would contain the true population parameter.

- It does not mean there is a 95% probability that the true value lies within one specific interval.

> How Confidence Intervals Are Used in Statistical Interpretation:

  1. Estimate Population Parameters: Provides a range instead of a single value, offering more information about the precision of the estimate.

  2. Assess Statistical Significance:

    - If a confidence interval for a difference in means or proportions does not include 0, it suggests a statistically significant difference.

    - If a confidence interval for a single mean does not include a hypothesized value, it may lead to rejecting the null hypothesis.

  3. Measure Reliability: Narrow intervals indicate more precise estimates; wide intervals suggest more uncertainty.

- Confidence interval gives a likely range for a population parameter based on sample data.

- It is a fundamental tool in inferential statistics for expressing uncertainty and interpreting the reliability and significance of statistical results.

### 15. What is the margin of error, and how does it affect the confidence interval?

- Margin of error (MOE) is a measure of the uncertainty or potential error in an estimate derived from sample data.

- It indicates how much the sample estimate (like a mean or proportion) could vary from the true population value.

- Margin of error is the amount added to and subtracted from the point estimate to create a confidence interval.

- It reflects the maximum expected difference between the sample estimate and the actual population parameter due to sampling variability.

- Formula => Margin of Error = Critical Value x Standard Error

- Critical Value: Based on the desired confidence level ex. Z or t value.

- Standard Error: Depends on the sample size and standard deviation.

> How It Affects the Confidence Interval:

  1. Larger Margin of Error → Wider Confidence Interval: More uncertainty, but more likely to contain the true parameter.

  2. Smaller Margin of Error → Narrower Confidence Interval: More precise estimate, but slightly less coverage if assumptions are off.

- Margin of error represents the uncertainty around a sample estimate.

- It directly determines the width of a confidence interval.

- Larger sample size or lower confidence level decreases the margin of error, resulting in a more precise confidence interval.

### 16. How is Bayes' Theorem used in statistics, and what is its significance?

- Bayes' Theorem is a foundational concept in probability and statistics that describes how to update the probability of a hypothesis as new evidence or information becomes available.

- Bayes' Theorem is stated as: P(A/B) = (P(B/A).P(A))/P(B)

- Where P(A/B) = Posterior probability - the probability of event A given that
B has occurred (updated belief), P(B/A) = Likelihood - the probability of observing B given that A is true, P(A) = Prior probability - the initial belief about A before observing B, P(B) = Marginal probability - the total probability of observing B, under all possible scenarios.

> Use in Statistics: In statistics, Bayes' Theorem is used in Bayesian inference, which is an approach to statistical inference that combines prior beliefs with current evidence.

  1. Parameter estimation: Updating beliefs about unknown parameters as data is collected.

  2. Decision making: Incorporating uncertainty and prior knowledge to make informed decisions.

  3. Hypothesis testing: Calculating the probability of a hypothesis being true after considering the data.

> Significance:

  1. Dynamic Updating: Allows incorporation of new data into existing beliefs, making it powerful in real-time and adaptive analyses.

  2. Subjective Probability: Unlike classical (frequentist) statistics, it accommodates prior knowledge and expert opinion.

  3. Versatility: Used across fields like medicine (diagnosis), machine learning (ex- Naive Bayes classifier), and finance (risk assessment).

- Bayes' Theorem provides a formal method for updating probabilities in light of new evidence, making it a cornerstone of Bayesian statistics and a powerful tool for reasoning under uncertainty.

### 17. What is the Chi-square distribution, and when is it used?

- Chi-square (χ²) distribution is a theoretical probability distribution that arises in statistics when dealing with squared deviations and variance-based calculations, especially in the context of categorical data.

- Chi-square distribution is the distribution of a sum of the squares of
k independent standard normal random variables: X^2 = Z1^2 + Z2^2+.....+ZK^2

- Where each Zi is a standard normal variable and k is the degrees of freedom.

> Key Characteristics:

  1. It is non-negative (values range from 0 to ∞).

  2. It is right-skewed, especially for small degrees of freedom, and becomes more symmetric as k increases.

  3. Its shape depends entirely on the degrees of freedom.

> When Is It Used?

  1. Testing for Independence: In contingency tables, to test whether two categorical variables are independent (Chi-square test of independence).


  2. Goodness-of-Fit Test: To test if an observed frequency distribution differs from a theoretical (expected) distribution.

  3. Test for Variance: To test hypotheses about the variance of a normally distributed population.

  4. Confidence Intervals for Variance: Used to construct confidence intervals for population variance or standard deviation.

- Chi-square distribution is crucial in inferential statistics, particularly for testing hypotheses about categorical data and variances.

- Its importance lies in providing a framework for evaluating how well observed data fit theoretical expectations.

### 18. What is the Chi-square goodness of fit test, and how is it applied?

- Chi-square goodness of fit test is a statistical test used to determine whether a set of observed categorical data matches an expected distribution.

- It helps assess how well a theoretical distribution fits the actual data.

- Test compares the observed frequencies in each category with the expected frequencies based on a specified distribution.

- Formula => x^2 = ∑(Oi - Ei)^2/Ei

- Where Oi : Observed frequency in category i, Ei = Expected frequency in category i, The sum is over all categories

> Assumptions:

  1. Data are counts (frequencies) in mutually exclusive categories.

  2. Expected frequencies should generally be 5 or more in each category.

  3. Categories are independent of each other.

> Steps in Applying the Test:

  1. State the hypotheses: H0 and Ha

  2. Calculate expected frequencies: Based on the assumed distribution.

  3. Compute the Chi-square statistic: Using the formula above.

  4. Determine degrees of freedom: df = number of categories - 1

  5. Compare to critical value (from Chi-square distribution) or compute the p-value.

  6. Make a decision: If the statistic exceeds the critical value or the p-value is below the significance level (e.g., 0.05), reject H0

- Chi-square goodness of fit test is used to assess whether an observed frequency distribution matches an expected one.

- It's widely applied in situations where we want to test whether data fit a specific categorical model or theoretical probability distribution.

### 19. What is the F-distribution, and when is it used in hypothesis testing?

- F-distribution is a theoretical probability distribution that arises frequently in the context of comparing variances and in ANOVA (Analysis of Variance).

- It is used in hypothesis testing when comparing ratios of variances from independent samples.

- F-distribution is defined as the ratio of two scaled chi-square distributions: F = (X1/d1)/(X2/d2)

- Where X1 and X2 are independent chi-square distributed variables, d1 and d2 are their respective degrees of freedom.

> Key Characteristics:

  1. Values are non-negative (starts at 0 and goes to ∞).

  2. It is right-skewed, especially when degrees of freedom are small.

  3. Its shape depends on two parameters: the degrees of freedom for the numerator and the denominator.

> When Is It Used in Hypothesis Testing?

  1. Testing Equality of Variances

  2. Analysis of Variance (ANOVA)

  3. Regression Analysis

- F-distribution is central to tests that involve comparing ratios of variances.

- It is most commonly used in ANOVA, regression, and variance comparison tests, making it a key tool in determining whether observed differences among groups or models are statistically significant.

### 20. What is an ANOVA test, and what are its assumptions?

- ANOVA (Analysis of Variance) test is a statistical method used to determine whether there are significant differences between the means of three or more independent groups.

- ANOVA examines how much of the total variability in the data can be attributed to:

  1. Between-group variation (differences among group means)

  2. Within-group variation (natural variation within each group)

- F = Variance between groups/Variance between groups

- high F-value suggests that at least one group mean is significantly different from the others.

> Types of ANOVA:

  1. One-way ANOVA: Tests the effect of a single factor on a dependent variable.

  2. Two-way ANOVA: Tests the effect of two factors and their interaction.

> Assumptions of ANOVA:

  1. Independence: Observations are independent of each other.

  2. Normality: The data within each group are approximately normally distributed.

  3. Homogeneity of variances (homoscedasticity): All groups have approximately equal variances.

- ANOVA is used to test for differences in means across multiple groups by analyzing variance.

- It is a fundamental tool in experimental design and statistical inference, but it relies on key assumptions like independence, normality, and equal variances for valid results.

### 21. What are the different types of ANOVA tests?

- There are several types of ANOVA (Analysis of Variance) tests, each designed for different experimental setups and research questions.

- All ANOVA tests aim to determine whether there are significant differences between group means, but they differ based on the number of factors (independent variables) and how they are structured.

  1. One-Way ANOVA - Purpose: Tests differences in means across three or more groups based on one independent variable (factor). Ex - Comparing average test scores across different teaching methods.

  2. Two-Way ANOVA - Purpose: Tests the effect of two independent variables on one dependent variable, and whether there is an interaction between them.2 Type of it - Without interaction: Tests only main effects, With interaction: Tests both main effects and their interaction. Ex- Studying the effect of diet and exercise (two factors) on weight loss.

  3. Repeated Measures ANOVA - Purpose: Used when the same subjects are measured under different conditions or over multiple time points.Ex- Measuring blood pressure of the same patients before, during, and after treatment.

  4. Mixed-Design ANOVA (Split-Plot ANOVA) - Purpose: Combines between-subjects factors (different groups) and within-subjects factors (repeated measures on the same subjects). Ex- Comparing treatment effects (between subjects) over multiple time points (within subjects).

  5. MANOVA (Multivariate ANOVA) - Purpose: An extension of ANOVA that tests for differences in two or more dependent variables simultaneously. Ex- Studying how different teaching methods affect both math and reading scores.

- So as summary there are different types of ANOVA are tailored to different experimental designs:

  1. One-way ANOVA: One factor, multiple groups.

  2. Two-way ANOVA: Two factors, possibly with interaction.

  3. Repeated measures ANOVA: Same subjects over time or conditions.

  4. Mixed-design ANOVA: Combines both repeated and independent factors.

  5. MANOVA: Multiple dependent variables.

```python
'''
# 1. Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and
interpret the results.
'''

import numpy as np
from scipy.stats import norm

# --- Input values ---
sample_data = [51, 53, 54, 56, 52, 55, 57, 58, 54, 55]   # sample values
population_mean = 50                                     # known population mean
population_std = 5                                       # known population standard deviation
alpha = 0.05                                             # significance level

# --- Step 1: Compute sample statistics ---
sample_mean = np.mean(sample_data)
sample_size = len(sample_data)

# --- Step 2: Compute Z-statistic ---
z_score = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))

# --- Step 3: Calculate p-value (two-tailed test) ---
p_value = 2 * (1 - norm.cdf(abs(z_score)))

# --- Step 4: Print results ---
print("Sample Mean:", round(sample_mean, 2))
print("Z-Score:", round(z_score, 4))
print("P-Value:", round(p_value, 4))

# --- Step 5: Conclusion ---
if p_value < alpha:
    print(f"Reject the null hypothesis at alpha = {alpha}. The sample mean is significantly different from the population mean.")
else:
    print(f"Fail to reject the null hypothesis at alpha = {alpha}. No significant difference found.")
```

```python
'''
# 4. Perform a two-tailed Z-test using Python and visualize the decision region on a plot.
'''

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# --- Step 1: Define inputs ---
sample_data = [102, 98, 101, 100, 99, 97, 103, 96, 100, 104]
population_mean = 100
population_std = 5
alpha = 0.05

# --- Step 2: Calculate sample statistics ---
sample_mean = np.mean(sample_data)
n = len(sample_data)
z_score = (sample_mean - population_mean) / (population_std / np.sqrt(n))
p_value = 2 * (1 - norm.cdf(abs(z_score)))

# --- Step 3: Decision logic ---
critical_z = norm.ppf(1 - alpha / 2)  # Two-tailed test

# --- Step 4: Print results ---
print("Sample Mean:", round(sample_mean, 2))
print("Z-Score:", round(z_score, 4))
print("P-Value:", round(p_value, 4))
print("Critical Z-Value:", round(critical_z, 4))

if abs(z_score) > critical_z:
    print(f"Reject the null hypothesis at alpha = {alpha}.")
else:
    print(f"Fail to reject the null hypothesis at alpha = {alpha}.")

# --- Step 5: Plot the decision region ---
x = np.linspace(-4, 4, 1000)
y = norm.pdf(x, 0, 1)

plt.figure(figsize=(10, 6))
plt.plot(x, y, label="Standard Normal Distribution")

# Shade rejection regions
plt.fill_between(x, y, where=(x <= -critical_z), color='red', alpha=0.5, label='Rejection Region (Left)')
plt.fill_between(x, y, where=(x >= critical_z), color='red', alpha=0.5, label='Rejection Region (Right)')

# Plot z-score
plt.axvline(z_score, color='blue', linestyle='--', linewidth=2, label=f'Z-Score = {z_score:.2f}')

# Plot critical values
plt.axvline(-critical_z, color='black', linestyle=':', linewidth=1, label=f'-Z Critical = {-critical_z:.2f}')
plt.axvline(critical_z, color='black', linestyle=':', linewidth=1, label=f'+Z Critical = {critical_z:.2f}')

plt.title("Two-Tailed Z-Test Decision Region")
plt.xlabel("Z")
plt.ylabel("Probability Density")
plt.legend()
plt.grid(True)
plt.show()
```

```python
'''
# 5. Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing.
'''

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def visualize_errors(mu0=100, mu1=105, sigma=10, n=30, alpha=0.05, tail='right'):
    """
    Visualize Type I and Type II errors for a Z-test in a hypothesis testing setup.

    Parameters:
    - mu0: Population mean under H0
    - mu1: True mean under H1
    - sigma: Population standard deviation
    - n: Sample size
    - alpha: Significance level
    - tail: 'right', 'left', or 'two' for one- or two-tailed test
    """

    # Standard error
    se = sigma / np.sqrt(n)

    # Critical values and regions
    if tail == 'right':
        z_crit = norm.ppf(1 - alpha)
        x_crit = mu0 + z_crit * se
        beta = norm.cdf((x_crit - mu1) / se)
    elif tail == 'left':
        z_crit = norm.ppf(alpha)
        x_crit = mu0 + z_crit * se
        beta = 1 - norm.cdf((x_crit - mu1) / se)
    elif tail == 'two':
        z_crit = norm.ppf(1 - alpha / 2)
        x_crit_low = mu0 - z_crit * se
        x_crit_high = mu0 + z_crit * se
        beta = norm.cdf((x_crit_high - mu1) / se) - norm.cdf((x_crit_low - mu1) / se)
    else:
        raise ValueError("tail must be 'left', 'right', or 'two'")

    # Plot range
    x = np.linspace(mu0 - 4*se, mu1 + 4*se, 1000)
    y_h0 = norm.pdf(x, mu0, se)
    y_h1 = norm.pdf(x, mu1, se)

    plt.figure(figsize=(10, 6))
    plt.plot(x, y_h0, label='H0 Distribution (μ₀)', color='black')
    plt.plot(x, y_h1, label='H1 Distribution (μ₁)', color='blue')

    # Shade Type I Error (alpha)
    if tail == 'right':
        plt.fill_between(x, y_h0, where=(x >= x_crit), color='red', alpha=0.4, label='Type I Error (α)')
        plt.fill_between(x, y_h1, where=(x < x_crit), color='blue', alpha=0.3, label='Type II Error (β)')
    elif tail == 'left':
        plt.fill_between(x, y_h0, where=(x <= x_crit), color='red', alpha=0.4, label='Type I Error (α)')
        plt.fill_between(x, y_h1, where=(x > x_crit), color='blue', alpha=0.3, label='Type II Error (β)')
    else:  # two-tailed
        plt.fill_between(x, y_h0, where=(x <= x_crit_low) | (x >= x_crit_high), color='red', alpha=0.4, label='Type I Error (α)')
        plt.fill_between(x, y_h1, where=(x > x_crit_low) & (x < x_crit_high), color='blue', alpha=0.3, label='Type II Error (β)')

    plt.title('Visualization of Type I and Type II Errors')
    plt.xlabel('Sample Mean')
    plt.ylabel('Probability Density')
    plt.legend()
    plt.grid(True)
    plt.show()

    # Print results
    print(f"Type I Error (α): {alpha}")
    print(f"Type II Error (β): {beta:.4f}")
    print(f"Power of the test (1 - β): {1 - beta:.4f}")

# --- Example use ---
visualize_errors(mu0=100, mu1=105, sigma=10, n=30, alpha=0.05, tail='right')
```

```python
'''
# 6. Write a Python program to perform an independent T-test and interpret the results.
'''

import numpy as np
from scipy.stats import ttest_ind

# --- Step 1: Simulate two independent samples ---
np.random.seed(42)  # for reproducibility

# Sample A (ex- Group 1)
group1 = np.random.normal(loc=100, scale=10, size=30)

# Sample B (ex- Group 2)
group2 = np.random.normal(loc=105, scale=12, size=30)

# --- Step 2: Perform independent T-test ---
t_stat, p_value = ttest_ind(group1, group2, equal_var=False)  # Welch's t-test (recommended if variances may differ)

# --- Step 3: Set significance level ---
alpha = 0.05

# --- Step 4: Print results ---
print("Group 1 Mean:", round(np.mean(group1), 2))
print("Group 2 Mean:", round(np.mean(group2), 2))
print("T-Statistic:", round(t_stat, 4))
print("P-Value:", round(p_value, 4))

# --- Step 5: Interpret the result ---
if p_value < alpha:
    print(f"\nResult: Reject the null hypothesis at alpha = {alpha}")
    print("Conclusion: There is a statistically significant difference between the group means.")
else:
    print(f"\nResult: Fail to reject the null hypothesis at alpha = {alpha}")
    print("Conclusion: There is no statistically significant difference between the group means.")
```

```python
'''
# 9. Write a Python function to calculate the confidence interval for a sample mean and explain its significance.
'''

import numpy as np
from scipy.stats import t

def confidence_interval(sample, confidence=0.95):
    """
    Calculate the confidence interval for the mean of a sample using the t-distribution.

    Parameters:
    - sample (list or np.array): Numeric sample data.
    - confidence (float): Confidence level (e.g., 0.95 for 95%).

    Returns:
    - (lower_bound, upper_bound): Confidence interval tuple.
    """
    sample = np.array(sample)
    n = len(sample)
    mean = np.mean(sample)
    std_err = np.std(sample, ddof=1) / np.sqrt(n)

    t_critical = t.ppf((1 + confidence) / 2, df=n - 1)
    margin_of_error = t_critical * std_err

    lower_bound = mean - margin_of_error
    upper_bound = mean + margin_of_error

    print(f"Sample Mean: {mean:.2f}")
    print(f"Confidence Level: {confidence*100:.0f}%")
    print(f"Margin of Error: ±{margin_of_error:.2f}")
    print(f"Confidence Interval: ({lower_bound:.2f}, {upper_bound:.2f})")

    return (lower_bound, upper_bound)

# --- Example usage ---
np.random.seed(0)
sample_data = np.random.normal(loc=50, scale=5, size=30)
confidence_interval(sample_data, confidence=0.95)
```

```python
'''
# 10. Write a Python program to calculate the margin of error for a given confidence level using sample data.
'''

import numpy as np
from scipy.stats import t

def margin_of_error(sample, confidence=0.95):
    """
    Calculate the margin of error for a sample mean using the t-distribution.

    Parameters:
    - sample: list or np.array of sample values
    - confidence: confidence level (default is 0.95 for 95%)

    Returns:
    - margin of error (float)
    """
    sample = np.array(sample)
    n = len(sample)
    sample_std = np.std(sample, ddof=1)
    std_err = sample_std / np.sqrt(n)

    t_crit = t.ppf((1 + confidence) / 2, df=n - 1)
    margin_error = t_crit * std_err

    print(f"Sample Size (n): {n}")
    print(f"Sample Mean: {np.mean(sample):.2f}")
    print(f"Sample Standard Deviation: {sample_std:.2f}")
    print(f"Confidence Level: {confidence*100:.0f}%")
    print(f"Margin of Error: ±{margin_error:.2f}")

    return margin_error

# --- Example Usage ---
np.random.seed(1)
sample_data = np.random.normal(loc=100, scale=15, size=25)
margin_of_error(sample_data, confidence=0.95)
```

```python
'''
# 11. Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process.
'''

def bayesian_update(prior_H, likelihood_E_given_H, likelihood_E_given_not_H):
    """
    Perform Bayesian inference using Bayes' Theorem.

    Parameters:
    - prior_H: Prior probability of hypothesis H
    - likelihood_E_given_H: P(Evidence | Hypothesis)
    - likelihood_E_given_not_H: P(Evidence | not Hypothesis)

    Returns:
    - posterior: Updated probability P(H | E)
    """
    prior_not_H = 1 - prior_H

    # Bayes' Theorem
    numerator = likelihood_E_given_H * prior_H
    denominator = numerator + (likelihood_E_given_not_H * prior_not_H)
    posterior = numerator / denominator

    print(f"Prior (P(H)): {prior_H:.2f}")
    print(f"Likelihood (P(E|H)): {likelihood_E_given_H:.2f}")
    print(f"Likelihood (P(E|¬H)): {likelihood_E_given_not_H:.2f}")
    print(f"Posterior (P(H|E)): {posterior:.4f}")

    return posterior

# --- Example usage ---
# Scenario: Medical test
# H = Has disease
# E = Positive test result
prior_H = 0.01  # 1% of population has the disease
likelihood_E_given_H = 0.99  # 99% sensitive test (true positive)
likelihood_E_given_not_H = 0.05  # 5% false positive rate

bayesian_update(prior_H, likelihood_E_given_H, likelihood_E_given_not_H)
```

```python
'''
# 12. Perform a Chi-square test for independence between two categorical variables in Python.
'''

import pandas as pd
import numpy as np
from scipy.stats import chi2_contingency

# --- Step 1: Simulate categorical data ---
np.random.seed(42)

# Example: Gender (Male/Female) and Preference (Like/Dislike)
gender = np.random.choice(['Male', 'Female'], size=100)
preference = np.random.choice(['Like', 'Dislike'], size=100)

# Create a DataFrame
data = pd.DataFrame({'Gender': gender, 'Preference': preference})

# --- Step 2: Create a contingency table ---
contingency_table = pd.crosstab(data['Gender'], data['Preference'])
print("Contingency Table:\n", contingency_table)

# --- Step 3: Perform Chi-square test for independence ---
chi2, p, dof, expected = chi2_contingency(contingency_table)

print(f"\nChi-square Statistic: {chi2:.4f}")
print(f"Degrees of Freedom: {dof}")
print(f"P-value: {p:.4f}")

# --- Step 4: Interpretation ---
alpha = 0.05
if p < alpha:
    print("\nResult: Reject the null hypothesis.")
    print("Conclusion: There is a significant association between Gender and Preference.")
else:
    print("\nResult: Fail to reject the null hypothesis.")
    print("Conclusion: No significant association between Gender and Preference.")
```

```python
'''
# 16. Implement an F-test using Python to compare the variances of two random samples.
'''

import numpy as np
from scipy.stats import f

# --- Step 1: Simulate two random samples ---
np.random.seed(42)
sample1 = np.random.normal(loc=50, scale=10, size=30)  # std dev 10
sample2 = np.random.normal(loc=50, scale=15, size=35)  # std dev 15

# --- Step 2: Calculate sample variances ---
var1 = np.var(sample1, ddof=1)
var2 = np.var(sample2, ddof=1)

# --- Step 3: Calculate F-statistic (ratio of variances) ---
# Larger variance as numerator to ensure F >= 1
if var1 > var2:
    F = var1 / var2
    dfn = len(sample1) - 1  # degrees of freedom numerator
    dfd = len(sample2) - 1  # degrees of freedom denominator
else:
    F = var2 / var1
    dfn = len(sample2) - 1
    dfd = len(sample1) - 1

# --- Step 4: Calculate p-value for two-tailed test ---
p_value = 2 * min(f.cdf(F, dfn, dfd), 1 - f.cdf(F, dfn, dfd))

print(f"Variance Sample 1: {var1:.2f}")
print(f"Variance Sample 2: {var2:.2f}")
print(f"F-Statistic: {F:.4f}")
print(f"Degrees of Freedom: numerator={dfn}, denominator={dfd}")
print(f"P-value: {p_value:.4f}")

# --- Step 5: Interpret results ---
alpha = 0.05
if p_value < alpha:
    print("Result: Reject the null hypothesis.")
    print("Conclusion: Variances are significantly different.")
else:
    print("Result: Fail to reject the null hypothesis.")
    print("Conclusion: No significant difference in variances.")
```

```python
'''
# 17. Write a Python program to perform an ANOVA test to compare means between multiple groups and
interpret the results.
'''

import numpy as np
from scipy.stats import f_oneway

# --- Step 1: Simulate data for 3 groups ---
np.random.seed(0)
group1 = np.random.normal(loc=20, scale=5, size=30)
group2 = np.random.normal(loc=22, scale=5, size=30)
group3 = np.random.normal(loc=19, scale=5, size=30)

# --- Step 2: Perform one-way ANOVA ---
F_statistic, p_value = f_oneway(group1, group2, group3)

print(f"ANOVA F-statistic: {F_statistic:.4f}")
print(f"P-value: {p_value:.4f}")

# --- Step 3: Interpretation ---
alpha = 0.05
if p_value < alpha:
    print("Result: Reject the null hypothesis.")
    print("Conclusion: At least one group mean is significantly different.")
else:
    print("Result: Fail to reject the null hypothesis.")
    print("Conclusion: No significant difference between group means.")
```

```python
'''
# 18. Perform a one-way ANOVA test using Python to compare the means of different groups and plot the results.
'''

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import f_oneway

# --- Step 1: Simulate data for 4 groups ---
np.random.seed(42)
group1 = np.random.normal(loc=50, scale=5, size=30)
group2 = np.random.normal(loc=52, scale=5, size=30)
group3 = np.random.normal(loc=48, scale=5, size=30)
group4 = np.random.normal(loc=55, scale=5, size=30)

groups = [group1, group2, group3, group4]

# --- Step 2: Perform one-way ANOVA ---
F_statistic, p_value = f_oneway(*groups)

print(f"ANOVA F-statistic: {F_statistic:.4f}")
print(f"P-value: {p_value:.4f}")

alpha = 0.05
if p_value < alpha:
    print("Result: Reject the null hypothesis.")
    print("Conclusion: At least one group mean is significantly different.")
else:
    print("Result: Fail to reject the null hypothesis.")
    print("Conclusion: No significant difference between group means.")

# --- Step 3: Visualize with boxplots ---
plt.figure(figsize=(8,6))
plt.boxplot(groups, labels=['Group 1', 'Group 2', 'Group 3', 'Group 4'])
plt.title('Group Distributions')
plt.ylabel('Values')
plt.grid(axis='y')
plt.show()
```

```python
'''
# 19. Write a Python function to check the assumptions (normality, independence, and equal variance) for ANOVA.
'''

import numpy as np
import scipy.stats as stats

def check_anova_assumptions(groups):
    """
    Check ANOVA assumptions:
    1. Normality per group (Shapiro-Wilk test)
    2. Equal variances (Levene's test)
    3. Independence (reminder: needs design check)

    Parameters:
    groups : list of arrays or lists (each group's data)

    Prints test results and interpretations.
    """
    alpha = 0.05
    print("Checking ANOVA assumptions...\n")

    # 1. Normality test (Shapiro-Wilk) per group
    print("1. Normality Test (Shapiro-Wilk) per group:")
    normality_pass = True
    for i, group in enumerate(groups, 1):
        stat, p = stats.shapiro(group)
        print(f"  Group {i}: W={stat:.4f}, p-value={p:.4f}", end=' ')
        if p > alpha:
            print("- Normality assumed")
        else:
            print("- Normality rejected")
            normality_pass = False

    # 2. Equal variances test (Levene's test)
    stat, p = stats.levene(*groups)
    print(f"\n2. Equal Variance Test (Levene's test): W={stat:.4f}, p-value={p:.4f}", end=' ')
    if p > alpha:
        print("- Equal variances assumed")
        equal_variance_pass = True
    else:
        print("- Equal variances rejected")
        equal_variance_pass = False

    # 3. Independence assumption
    print("\n3. Independence assumption:")
    print("   - Independence must be ensured by the study design and data collection process.")
    print("   - It cannot be tested statistically here.\n")

    # Summary
    print("Summary:")
    if normality_pass and equal_variance_pass:
        print("All assumptions satisfied: You can proceed with ANOVA.")
    else:
        print("Some assumptions failed: Consider data transformations or non-parametric tests.")


# --- Example usage ---
import matplotlib.pyplot as plt

np.random.seed(123)
g1 = np.random.normal(20, 5, 30)
g2 = np.random.normal(22, 5, 30)
g3 = np.random.normal(19, 8, 30)

groups = [g1, g2, g3]

check_anova_assumptions(groups)
```

```python
'''
# 20. Perform a two-way ANOVA test using Python to study the interaction between two factors and visualize the
results.
'''

import numpy as np
import pandas as pd
import statsmodels.api as sm
from statsmodels.formula.api import ols
import matplotlib.pyplot as plt
import seaborn as sns

# --- Step 1: Simulate data ---

np.random.seed(42)
levels_A = ['Low', 'High']
levels_B = ['Type1', 'Type2', 'Type3']

# Create a balanced design with 10 samples per group
data = []
for A in levels_A:
    for B in levels_B:
        # Mean depends on factors + interaction effect
        if A == 'Low' and B == 'Type1':
            mean = 10
        elif A == 'Low' and B == 'Type2':
            mean = 12
        elif A == 'Low' and B == 'Type3':
            mean = 14
        elif A == 'High' and B == 'Type1':
            mean = 15
        elif A == 'High' and B == 'Type2':
            mean = 18
        else:
            mean = 16

        # Simulate 10 observations per cell
        samples = np.random.normal(loc=mean, scale=2, size=10)
        for s in samples:
            data.append({'FactorA': A, 'FactorB': B, 'Value': s})

df = pd.DataFrame(data)

# --- Step 2: Two-way ANOVA with interaction ---
model = ols('Value ~ C(FactorA) * C(FactorB)', data=df).fit()
anova_table = sm.stats.anova_lm(model, typ=2)
print("ANOVA Table:")
print(anova_table)

# --- Step 3: Visualize interaction plot ---
plt.figure(figsize=(8,6))
sns.pointplot(x='FactorB', y='Value', hue='FactorA', data=df,
              dodge=True, markers=['o', 's'], capsize=0.1, errwidth=1, palette='Set2')
plt.title('Interaction Plot: FactorA and FactorB')
plt.ylabel('Mean Value')
plt.grid(True)
plt.show()
```

```python
'''
# 21. Write a Python program to visualize the F-distribution and discuss its use in hypothesis testing.
'''

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import f

# Degrees of freedom pairs (df1, df2)
df_pairs = [(1, 10), (5, 10), (10, 10), (10, 20)]

x = np.linspace(0, 5, 500)

plt.figure(figsize=(10,6))
for dfn, dfd in df_pairs:
    plt.plot(x, f.pdf(x, dfn, dfd), label=f'df1={dfn}, df2={dfd}')

plt.title('F-distribution PDFs for Different Degrees of Freedom')
plt.xlabel('F value')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True)
plt.show()
```

```python
'''
# 22. Perform a one-way ANOVA test in Python and visualize the results with boxplots to compare group means.
'''

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import f_oneway

# --- Step 1: Simulate sample data for 3 groups ---
np.random.seed(123)
group1 = np.random.normal(loc=20, scale=4, size=30)
group2 = np.random.normal(loc=22, scale=4, size=30)
group3 = np.random.normal(loc=19, scale=4, size=30)

groups = [group1, group2, group3]

# --- Step 2: Perform one-way ANOVA ---
F_statistic, p_value = f_oneway(*groups)

print(f"ANOVA F-statistic: {F_statistic:.4f}")
print(f"P-value: {p_value:.4f}")

alpha = 0.05
if p_value < alpha:
    print("Result: Reject the null hypothesis.")
    print("Conclusion: At least one group mean is significantly different.")
else:
    print("Result: Fail to reject the null hypothesis.")
    print("Conclusion: No significant difference between group means.")

# --- Step 3: Visualize group distributions with boxplots ---
plt.figure(figsize=(8,6))
plt.boxplot(groups, labels=['Group 1', 'Group 2', 'Group 3'])
plt.title('Boxplots of Group Distributions')
plt.ylabel('Values')
plt.grid(axis='y')
plt.show()
```

```python
'''
# 23. Simulate random data from a normal distribution, then perform hypothesis testing to evaluate the means.
'''

import numpy as np
from scipy.stats import ttest_ind

# --- Step 1: Simulate random data ---
np.random.seed(42)
sample1 = np.random.normal(loc=50, scale=5, size=40)  # mean=50, std=5
sample2 = np.random.normal(loc=52, scale=5, size=40)  # mean=52, std=5

# --- Step 2: Perform two-sample independent t-test ---
t_stat, p_value = ttest_ind(sample1, sample2)

print(f"T-statistic: {t_stat:.4f}")
print(f"P-value: {p_value:.4f}")

# --- Step 3: Interpret results ---
alpha = 0.05
if p_value < alpha:
    print("Result: Reject the null hypothesis.")
    print("Conclusion: The means are significantly different.")
else:
    print("Result: Fail to reject the null hypothesis.")
    print("Conclusion: No significant difference between the means.")
```

```python
'''
# 24. Perform a hypothesis test for population variance using a Chi-square distribution and interpret the results.
'''

import numpy as np
from scipy.stats import chi2

# --- Step 1: Simulate sample data ---
np.random.seed(0)
sample = np.random.normal(loc=100, scale=15, size=30)  # mean=100, std=15

# Sample statistics
n = len(sample)
sample_var = np.var(sample, ddof=1)

# Hypothesized population variance
sigma0_squared = 225  # e.g., (15^2)

# --- Step 2: Calculate Chi-square test statistic ---
chi_stat = (n - 1) * sample_var / sigma0_squared

# Degrees of freedom
df = n - 1

# --- Step 3: Calculate two-tailed p-value ---
p_lower = chi2.cdf(chi_stat, df)
p_upper = 1 - p_lower
p_value = 2 * min(p_lower, p_upper)  # two-tailed

print(f"Sample variance: {sample_var:.2f}")
print(f"Chi-square statistic: {chi_stat:.4f}")
print(f"Degrees of freedom: {df}")
print(f"P-value (two-tailed): {p_value:.4f}")

# --- Step 4: Interpret results ---
alpha = 0.05
if p_value < alpha:
    print("Result: Reject the null hypothesis.")
    print("Conclusion: The population variance is significantly different from the hypothesized value.")
else:
    print("Result: Fail to reject the null hypothesis.")
    print("Conclusion: No significant difference from the hypothesized population variance.")
```

```python
'''
# 25. Write a Python script to perform a Z-test for comparing proportions between two datasets or groups.
'''

import numpy as np
from statsmodels.stats.proportion import proportions_ztest

# Example data:
# Group 1: successes and sample size
successes_1 = 45
n1 = 100

# Group 2: successes and sample size
successes_2 = 30
n2 = 80

# --- Perform two-proportion Z-test ---
count = np.array([successes_1, successes_2])
nobs = np.array([n1, n2])

stat, pval = proportions_ztest(count, nobs)

print(f"Z-statistic: {stat:.4f}")
print(f"P-value: {pval:.4f}")

alpha = 0.05
if pval < alpha:
    print("Result: Reject the null hypothesis.")
    print("Conclusion: The proportions are significantly different.")
else:
    print("Result: Fail to reject the null hypothesis.")
    print("Conclusion: No significant difference between the proportions.")
```

```python
'''
# 26. Implement an F-test for comparing the variances of two datasets, then interpret and visualize the results.
'''

import numpy as np
from scipy.stats import f
import matplotlib.pyplot as plt

# --- Step 1: Simulate two datasets ---
np.random.seed(42)
data1 = np.random.normal(loc=50, scale=5, size=40)  # variance around 25
data2 = np.random.normal(loc=50, scale=8, size=40)  # variance around 64

# --- Step 2: Calculate sample variances ---
var1 = np.var(data1, ddof=1)
var2 = np.var(data2, ddof=1)

# Calculate degrees of freedom
df1 = len(data1) - 1
df2 = len(data2) - 1

# --- Step 3: Calculate F-statistic ---
# F = larger variance / smaller variance
if var1 > var2:
    F_stat = var1 / var2
    dfn, dfd = df1, df2
else:
    F_stat = var2 / var1
    dfn, dfd = df2, df1

# --- Step 4: Calculate p-value for two-tailed test ---
p_value = 2 * min(f.cdf(F_stat, dfn, dfd), 1 - f.cdf(F_stat, dfn, dfd))

print(f"Variance data1: {var1:.2f}")
print(f"Variance data2: {var2:.2f}")
print(f"F-statistic: {F_stat:.4f}")
print(f"Degrees of freedom: numerator={dfn}, denominator={dfd}")
print(f"P-value (two-tailed): {p_value:.4f}")

alpha = 0.05
if p_value < alpha:
    print("Result: Reject the null hypothesis.")
    print("Conclusion: The variances are significantly different.")
else:
    print("Result: Fail to reject the null hypothesis.")
    print("Conclusion: No significant difference in variances.")

# --- Step 5: Visualization ---

# Boxplots of the two samples
plt.figure(figsize=(12,5))

plt.subplot(1, 2, 1)
plt.boxplot([data1, data2], labels=['Data1', 'Data2'])
plt.title('Boxplot of the Two Datasets')
plt.ylabel('Values')
plt.grid(axis='y')

# Plot F-distribution and mark F_stat
x = np.linspace(0, 5, 500)
y = f.pdf(x, dfn, dfd)

plt.subplot(1, 2, 2)
plt.plot(x, y, label=f'F-distribution\n(df1={dfn}, df2={dfd})')
plt.axvline(F_stat, color='red', linestyle='--', label=f'F-statistic = {F_stat:.2f}')
plt.fill_between(x, 0, y, alpha=0.2)
plt.title('F-distribution with Test Statistic')
plt.xlabel('F value')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
'''
# 27. Perform a Chi-square test for goodness of fit with simulated data and analyze the results.
'''

import numpy as np
from scipy.stats import chisquare

# --- Step 1: Simulate observed counts for categories ---
np.random.seed(123)
categories = ['A', 'B', 'C', 'D']
total_samples = 200

# Suppose true probabilities are:
true_probs = [0.25, 0.35, 0.30, 0.10]

# Generate observed counts by sampling
observed_counts = np.random.multinomial(total_samples, true_probs)
print("Observed counts:", dict(zip(categories, observed_counts)))

# --- Step 2: Define expected proportions (null hypothesis) ---
expected_probs = [0.25, 0.25, 0.25, 0.25]  # e.g., equal proportions
expected_counts = np.array(expected_probs) * total_samples
print("Expected counts:", dict(zip(categories, expected_counts.astype(int))))

# --- Step 3: Perform Chi-square goodness-of-fit test ---
chi2_stat, p_value = chisquare(f_obs=observed_counts, f_exp=expected_counts)

print(f"\nChi-square statistic: {chi2_stat:.4f}")
print(f"P-value: {p_value:.4f}")

# --- Step 4: Interpret results ---
alpha = 0.05
if p_value < alpha:
    print("Result: Reject the null hypothesis.")
    print("Conclusion: Observed distribution significantly differs from expected.")
else:
    print("Result: Fail to reject the null hypothesis.")
    print("Conclusion: No significant difference from the expected distribution.")
```
